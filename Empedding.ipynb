```python
#################################################################
# CELL 1: TẠO VÀ LƯU EMBEDDINGS (PHOBERT FINE-TUNED)
#################################################################
import pandas as pd
import numpy as np
import torch
from sentence_transformers import SentenceTransformer

# --- 1. Cài đặt ---
MODEL_NAME = "keepitreal/vietnamese-sbert"
DF_PATH = "dantri_qc_grouped.csv"
EMB_SAVE_PATH = "vietnamese_sbert_embeddings.npy"

# Kiểm tra GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Sử dụng thiết bị: {device}")

# --- 2. Tải dữ liệu ---
print(f"Đang tải dữ liệu từ: {DF_PATH}")
df = pd.read_csv(DF_PATH)
texts = df["clean_text"].astype(str).tolist()
print(f"Đã tải {len(texts)} tài liệu.")

# --- 3. Tải mô hình SBERT ---
print(f"Đang tải mô hình embedding: {MODEL_NAME}...")
model = SentenceTransformer(MODEL_NAME, device=device)

# --- 4. Tạo Embeddings (Bước nặng) ---
print("Bắt đầu tạo embeddings...")
embeddings = model.encode(
    texts,
    batch_size=128,  # Tăng/giảm batch_size tùy theo VRAM của bạn
    show_progress_bar=True,
    convert_to_numpy=True,
    device=device
)
print("Hoàn thành tạo embeddings. Shape:", embeddings.shape)

# --- 5. Lưu Embeddings ---
np.save(EMB_SAVE_PATH, embeddings)
print(f"ĐÃ LƯU EMBEDDINGS TẠI: {EMB_SAVE_PATH}")
print(" HOÀN THÀNH CELL 1 ")
```
